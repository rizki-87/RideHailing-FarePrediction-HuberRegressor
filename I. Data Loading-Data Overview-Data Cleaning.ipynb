# Ib. Data Cleaning

```python
# creating a new DataFrame called `df1` by making a copy of the existing DataFrame `df_main`
df1 = df_main.copy()
```

```python
# check missing value.
missing_values_per_column = df1.isna().sum()
print(missing_values_per_column)
```

It has been identified that the "price" column contains 55,095 missing values. This means that there are 55,095 entries in the "price" column that do not have a valid value.

```python
# calculate the total number of rows in the DataFrame
total_rows = len(df1)
# calculate the percentage of missing values per column
percentage_missing_per_column = (missing_values_per_column / total_rows) * 100

# print the percentage of missing values per column
print(percentage_missing_per_column)
```

It has been determined that there is a 7% missing value rate in the dataset. To address these missing values, we have decided to impute them using the mean value of the respective column. This means that for each missing value in a column, we will replace it with the mean value of that column.

```python
# handling missing value with "mean" imputation
mean_price = df1['price'].mean()
df1['price'].fillna(mean_price, inplace=True)
```

```python
# checking after handling
missing_values_per_column = df1.isna().sum()
print(missing_values_per_column)
```

```python
# calculate the total number of rows in the DataFrame
total_rows = len(df1)
# calculate the percentage of missing values per column
percentage_missing_per_column = (missing_values_per_column / total_rows) * 100

# print the percentage of missing values per column
print(percentage_missing_per_column)
```

Missing values have been successfully resolved.
